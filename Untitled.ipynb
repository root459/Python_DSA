#### Techniques to measure time complexity
1) Measuring time to execute
2) Counting operations involved
3) Abstract notion of order of growth

##### Measuring time
work for=>different time for different algo
**problem with this approach**
1) time varies if implementation changes
3) different machines differnt time
4) does not work for extremely small input
5) time varies for different inputs but cant establish a relationship

```python
# gives how much time required to run the program 
# not used in industry because time depends on machine specifications 
# changing small things in code leads to change in time.ex=:> replace for with while
import time 
start=time.time()
for i in range(1,5):
    for j in range(1,20):
        print(i,j)
print(time.time()-start)
    
```
